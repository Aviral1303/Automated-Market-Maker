# AMM Model Comparison

Compare swap outputs and price impact across Constant Product, Balancer, and StableSwap.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))

from core import ConstantProductAMM, BalancerAMM, StableSwapAMM
from decimal import Decimal

reserve_a, reserve_b = 1_000_000, 1_000_000
swap_amount = 10_000

models = {
    "Constant Product": ConstantProductAMM(reserve_a, reserve_b),
    "Balancer 80/20": BalancerAMM(reserve_a, reserve_b, weight_a=0.8),
    "StableSwap": StableSwapAMM(reserve_a, reserve_b, amplification=100),
}

for name, amm in models.items():
    if amm.total_supply == 0:
        amm.add_liquidity(reserve_a, reserve_b)
    r = amm.swap("A", swap_amount)
    print(f"{name}: out={r.amount_out:.2f}, impact={r.price_impact_bps} bps")
```
